```python
import pandas as pd

def usuarios_que_cumplen(orden="None"):

    arch = pd.read_csv("log_catedras.csv")
    
    filas = []

    for i, fila in arch.iterrows():
        if fila["Contexto del evento"] in [
            "Tarea: Actividad 1 por Python plus - PARTE I",
            "Tarea: Actividad 1 por Python plus - PARTE II"
        ]:
            filas.append({"Usuario afectado": fila["Usuario afectado"]})

    # armo el dataframe
    df = pd.DataFrame(filas)

    # agrupo y cuento
    accesos_por_usuario = df["Usuario afectado"].value_counts()
    resultado = pd.DataFrame({
        "Usuario": accesos_por_usuario.index,
        "Cantidad de accesos": accesos_por_usuario.values
    })

    if orden == "A":
        resultado = resultado.sort_values(by="Cantidad de accesos", ascending=True)
    else:
        resultado = resultado.sort_values(by="Cantidad de accesos", ascending=False)

    return resultado
```

```python
import pandas as pd

print("Los primeros 5 usuarios que  realizaron la Actividad 1 en orden descendente")
print (" ")
datos_desc = usuarios_que_cumplen()
print(datos_desc.head(5))

print (" ")

print("Los primeros 5 usuarios que  realizaron la Actividad 1 en orden ascendente:")
print (" ")
datos_asc = usuarios_que_cumplen("A")
print(datos_asc.head(5))

print (" ")
#saco los datos que me pide
total = datos_desc["Cantidad de accesos"].sum()
maximo = datos_desc["Cantidad de accesos"].max()
minimo = datos_desc["Cantidad de accesos"].min()
promedio = datos_desc["Cantidad de accesos"].mean()

print("INFORMACIÓN A MOSTRAR")
print(f"Total de accesos: {total}")
print(f"Máximo: {maximo}")
print(f"Mínimo: {minimo}")
print(f"Promedio: {promedio:.2f}")
```
