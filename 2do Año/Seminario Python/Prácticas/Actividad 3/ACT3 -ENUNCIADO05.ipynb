## Parte I - CONSIGNA A RESOLVER

**EJERCICIO 1.** Escribir una función que retorne los usuarios que nunca accedieron a las grabaciones de las clases de teoría. Esta función recibe un parámetro opcional que indica si los nombres de usuario se muestran en mayúsculas **("M")**, minúsculas **("m")** o como aparecen en el dataset **(None)**. Por defecto la función retorna la información tal como aparecen en el dataset. 

**NOTA:** la información sobre el acceso a la grabación de la clase se obtiene  verificando que la columna "Nombre evento"  contenga el texto "Grabación vista" **y** la columna "Contexto del evento" contenga el  texto "BigBlueButton: Consultas por VC".

**EJERCICIO 2.** Utilizando la función anterior, escribir un programa que ingrese desde el teclado un menú de opciones que permita mostrar los usuarios que nunca accedieron a  las grabaciones de las clases de teoría con los nombres en mayúsculas, minúsculas o tal como aparecen en el dataset. 

## Solución

### EJERCICIO 1

```python
def usuariosQueNoAccedieron(csv_reader, formato=None):

    usuarios = set()
    usuariosConAcceso = set()

    
    for f in csv_reader:
        usuario = f['Nombre usuario']
        usuarios.add(usuario)
        
        if ('Grabación vista' in f['Nombre evento']) and ('BigBlueButton: Consultas por VC' in f['Contexto del evento']):
            usuariosConAcceso.add(usuario)

    lista = usuarios - usuariosConAcceso

    if formato == 'M':
        usuarios_sin_acceso = list(map(lambda x: x.upper(), lista))
    elif formato == 'm':
        usuarios_sin_acceso = list(map(lambda x: x.lower(), lista))
    else:
        usuarios_sin_acceso = list(lista)
    
    return lista
```

```python
import csv
from pathlib import Path

pathFile = Path('archivos') / 'logs_catedra.csv'

with open(pathFile, encoding='utf-8') as f:
    csv_reader = list(csv.DictReader(f))

    print("Usuarios que nunca accedieron a las grabaciones en el formato original: ")
    usuariosTotales = usuariosQueNoAccedieron (csv_reader, None)
    for user in usuariosTotales:
        print(user)
    
    print("Los usuarios en mayúsculas: ")

    usuariosMayuscula = usuariosQueNoAccedieron(csv_reader, 'M')
    for usuario in usuariosMayuscula:
        print(user)
    
    print("Los usuarios en minúsculas: ")

    usuariosMinuscula = usuariosQueNoAccedieron (csv_reader, 'm')
    for user in usuariosMinuscula:
        print(user)
```
