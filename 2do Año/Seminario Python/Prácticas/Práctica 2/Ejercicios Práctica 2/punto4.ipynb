```python
usuario = input("Ingrese el nombre de usuario: ")

if len(usuario) < 5:
    print ("❌ Debe tener al menos 5 caracteres.")
if not any(c.isdigit() for c in usuario):
    print ("❌ Debe contener al menos un número.")
if not any(c.isupper() for c in usuario):
    print ("❌ Debe contener al menos una letra mayúscula.")
if not usuario.isalnum():
    print ("❌ Solo puede contener letras y números.")
```
